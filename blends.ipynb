```python
from ogb.utils import smiles2graph
import numpy as np
from torch_geometric.data import InMemoryDataset, download_url, Data
import torch
import tqdm

def to_torch(graph):
    tensor_keys = ["edge_index", 'edge_feat', 'node_feat']
    for key in tensor_keys:
        graph[key] = torch.tensor(graph[key])
    return Data(x=graph["node_feat"].float(),edge_attr=graph["edge_feat"],edge_index=graph["edge_index"])

graph_data = dict()
for smiles in tqdm.tqdm(smiles_df["SMILES"]):
    graph_data[smiles] = to_torch(smiles2graph(smiles))
len(graph_data), next(iter(graph_data.items()))
```

```python
from torch_geometric.loader import DataLoader

def combine_graphs(mixture):
    graphs = [graph_data[cid_to_smiles[cid]["SMILES"]] for cid in mixture["cids"]]
    combined_batch = next(iter(DataLoader(graphs, batch_size=len(graphs))))
    return Data(x=combined_batch.x,edge_attr=combined_batch.edge_attr,edge_index=combined_batch.edge_index)

class PairData(Data):
    def __inc__(self, key, value, *args, **kwargs):
        if key == 'edge_index_s':
            return self.x_s.size(0)
        if key == 'edge_index_t':
            return self.x_t.size(0)
        return super().__inc__(key, value, *args, **kwargs)

def make_mixture_data(results,mixture_df,mixture_col,use_dataset,use_y):
    label_to_data = dict()
    for mixture in results:
        label_to_data[mixture["label"]] = combine_graphs(mixture)

    mixture_data = []
    for _,row in mixture_df.iterrows():
        label1 = (row["Dataset"] + "_" if use_dataset else "") + str(int(row[mixture_col+"1"]))
        graph1 = label_to_data[label1]
        label2 = (row["Dataset"] + "_"  if use_dataset else "") + str(int(row[mixture_col+"2"]))
        graph2 = label_to_data[label2]
        
        pair_data = PairData(label=f"{label1}/{label2}",x_s=graph1.x,edge_attr_s=graph1.edge_attr,edge_index_s=graph1.edge_index,x_t=graph2.x,edge_attr_t=graph2.edge_attr,edge_index_t=graph2.edge_index)
        if use_y:
            pair_data.y = row["Experimental Values"]
        mixture_data.append(pair_data)
        
    return mixture_data

# Wow what a crazy dataset :(
train_mixture_data = make_mixture_data(train_results,train_mixture_df,"Mixture ",use_dataset=True,use_y=True)
len(train_mixture_data), train_mixture_data[0]
```
