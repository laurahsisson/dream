```python
import pandas as pd

df_A = pd.read_csv("Mixtures/mixture_a_600.csv")
df_A
```

```python
label_a_to_str = {
    label_a: f"A_{i+1}" for i, label_a in enumerate(df_A["Mixture Label"].tolist())
}
label_a_to_str
```

```python
df_B = pd.read_csv("Mixtures/mixture_b.csv")
df_B
```

```python
label_b_to_str = {label_b: f"B_{i+1}" for i, label_b in enumerate(df_B.index.tolist())}
label_b_to_str
```

```python
distance_df = df_B[["Mixture A", "Distance"]].copy()
distance_df["Mixture A"] = distance_df["Mixture A"].map(label_a_to_str)
distance_df["Mixture B"] = distance_df.index.map(label_b_to_str)
distance_df["POM Pearson"] = distance_df["Distance"]
distance_df = distance_df[["Mixture A", "Mixture B", "POM Pearson"]]
display(distance_df)
distance_df.to_csv("Output/Mixture_Distance.csv")
```

```python
definition_A = df_A.copy()
definition_A["Mixture Label"] = definition_A["Mixture Label"].map(label_a_to_str)
definition_A = definition_A.drop(columns=["Unnamed: 0", "Dataset"])
definition_A
```

```python
definition_B = df_B.copy()
definition_B["Mixture Label"] = distance_df.index.map(label_b_to_str)
definition_B = definition_B.drop(columns=["Unnamed: 0", "Mixture A", "Distance"])
definition_B["CID"] = definition_B["CID.0"]
definition_B = definition_B.drop(columns=["CID.0"])
definition_B
```

```python
combined_definitions = pd.concat([definition_A, definition_B])
display(combined_definitions)
combined_definitions.to_csv("Output/Mixture_Definitions.csv")
```
